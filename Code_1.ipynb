```python
import sklearn
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.describe()
```

```python
train
```

```python
# define min max scaler
scaler = MinMaxScaler()
# transform data
cols = ['y'] + [f'{i}' for i in range (50)]
train[cols] = scaler.fit_transform(train[cols])

train
```

```python
targets = train['y']
```

```python
train = train.drop(columns=['y','w'])
```

```python
features_and_targets = pd.concat([train, targets], axis=1)
```

```python
corr_matrix = features_and_targets.corr()
corr_matrix['1']
```

```python
train_set, test_set = train_test_split(features_and_targets,test_size=0.2,train_size=0.8)
```

```python
train_set_features = train_set.drop(columns="1")
train_set_targets = train_set["1"]

lin_reg = LinearRegression().fit(train_set_features,train_set_targets)
```

```python
test_set_features = test_set.drop(columns="1")
test_set_targets = test_set["1"]

train_predictions = lin_reg.predict(train_set_features)
test_predictions = lin_reg.predict(test_set_features)

# We can check what MSE we get for random predictions. 
# If this is not significantly worse than for our true predictions, something is wrong!

from numpy.random import permutation

print("Test: ", np.sqrt(mean_squared_error(test_set_targets,test_predictions)))
print("Test_Random: ", np.sqrt(mean_squared_error(permutation(test_set_targets),test_predictions)))
print("Train: ", np.sqrt(mean_squared_error(train_set_targets,train_predictions)))
print("Train_Random: ", np.sqrt(mean_squared_error(permutation(train_set_targets),train_predictions)))
```
